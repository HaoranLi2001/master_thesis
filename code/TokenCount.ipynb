```python
import json
import matplotlib.pyplot as plt
import numpy as np

# 1. 加载保存好的 JSON 文件
filename = 'reasonlite_train1.json'
with open(filename, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 2. 筛选出 ReasonLite 的数据并计算 response 长度
# 我们在之前的步骤中给 ReasonLite 数据打上了 "type": "ReasonLite_Medium" 的标签
lengths = [len(item['response']) for item in data if "ReasonLite" in item.get('type', '')]

if not lengths:
    print("未找到类型为 'ReasonLite' 的数据，请检查 JSON 文件中的 type 字段。")
else:
    # 3. 计算基本统计指标
    avg_len = np.mean(lengths)
    max_len = np.max(lengths)
    min_len = np.min(lengths)
    median_len = np.median(lengths)

    print(f"--- ReasonLite Data Statistics (Total: {len(lengths)} items) ---")
    print(f"Average Length: {avg_len:.2f}")
    print(f"Maximum Length: {max_len}")
    print(f"Minimum Length: {min_len}")
    print(f"Median Length: {median_len}")
```
